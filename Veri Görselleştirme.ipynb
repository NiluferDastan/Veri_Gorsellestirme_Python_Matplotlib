### 4.4 plt.subplot2grid() fonksiyonu

Izgaraların içerisine istenilen boyutlarda grafik eklemek için subplot2grid() fonksiyonu kullanılabilir.Bu fonksiyon sırasıyla shape, location, rowspan,colspan parametrelerini alır.
shape: satır ve sütun değerlerini alır.1.girdi satır sayısını 2.girdi sütun sayısını gösterir.
loc: 2 tamsayı alır.Izgara içerisinde nereye yerleştirileceğini belirler.
rowspan: Sağ tarafa doğru ne kadar genişleyeceğini belirler.
colspan: Aşağı doğru ne kadar genişleyeceğini belirler.

*Izgaraların içerisine istenilen boyutlarda grafik eklemek için subplot2grid() fonksiyonu kullanılabilir.Bu fonksiyon sırasıyla shape, location, rowspan,colspan parametrelerini alır.


**shape:** satır ve sütun değerlerini alır.1.girdi satır sayısını 2.girdi sütun sayısını gösterir.

**loc:** 2 tamsayı alır.Izgara içerisinde nereye yerleştirileceğini belirler.

**rowspan:** Sağ tarafa doğru ne kadar genişleyeceğini belirler.

**colspan:** Aşağı doğru ne kadar genişleyeceğini belirler.*

```python
a1 = plt.subplot2grid((3,3),(0,0),colspan=2)
a2 = plt.subplot2grid((3,3),(0,2),rowspan=3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan=2, colspan=2)
x=np.arange(1,10)
a2.plot(x,x*x)
a2.set_title('square')
a1.plot(x,np.exp(x))
a1.set_title('exp')
a3.plot(x,np.log(x))
a3.set_title('log')
plt.tight_layout()
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

t=np.arange(0.0, 5.0, 0.01)
s=np.cos(2*np.pi*t)
line, = ax.plot(t,s,lw=2)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

t=np.arange(0.0, 5.0, 0.01)
s=np.cos(2*np.pi*t)
line, = ax.plot(t,s,lw=2)

ax.annotate('lokal max', xy=(2,1), xytext=(3,1.5),
           arrowprops=dict(facecolor='black',shrink=0.05),
           )
ax.set_ylim(-2,2)
plt.show()
```

### 8.1 *savefig() fonksiyonu*

*Matplotlib kütüphanesinde grafikleri resim olarak kaydetmek, grafikleri görüntülemek kadar kolaydır. Savefig() işlevi, grafikleri jpeg ve png gibi formatlarda kaydetmek için kullanılabilir. Grafiğin yolu ve adı savefig() fonksiyonuna parametre olarak verilmelidir. Bu parametre fname olarak verilmektedir. Fname dışında kullanılabilecek diğer bazı parametreler aşağıda verilmiştir.*

+ dpi=None GÖrüntünün çözünürlüğü, örneğin 300 dpi(inç başına düşen nokta sayısı)*

+ transparent=False. True değeri verildiğinde grafiğin arka planı transparan haline gelir.
+ bbox_inches=None. Grafik çevresindeki sınırlar ile ilgilidir.bbox_inches='tight' genellikle ideal sonuçlar verir.

*PNG formatı internet kullanımları ve çizim grafikleri için daha uygundur.
